```python
def imShow(path):
  import cv2
  import matplotlib.pyplot as plt
  %matplotlib inline

  image = cv2.imread(path)
  height, width = image.shape[:2]
  resized_image = cv2.resize(image,(3*width, 3*height), interpolation = cv2.INTER_CUBIC)

  fig = plt.gcf()
  fig.set_size_inches(18, 10)
  plt.axis("off")
  plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
  plt.show()
```

```python
import os
import yaml
from collections import defaultdict

# === 1. Đọc tên class từ data.yaml ===
yaml_path = '/content/drive/MyDrive/yolov5/ML/dataset_balanced/train_data/data.yaml'

with open(yaml_path, 'r') as f:
    data_yaml = yaml.safe_load(f)
class_names = data_yaml['names']

# === 2. Đường dẫn labels ===
label_dir_train = '/content/drive/MyDrive/yolov5/ML/dataset_balanced/train_data/labels/train'
label_dir_val = '/content/drive/MyDrive/yolov5/ML/dataset_balanced/train_data/labels/val'

def count_labels(label_dir):
    count_per_class = defaultdict(int)
    image_count = 0

    for label_file in os.listdir(label_dir):
        if label_file.endswith('.txt'):
            image_count += 1
            with open(os.path.join(label_dir, label_file), 'r') as f:
                lines = f.readlines()
                classes_in_image = set()
                for line in lines:
                    cls_id = int(line.split()[0])
                    classes_in_image.add(cls_id)
                for cls_id in classes_in_image:
                    count_per_class[cls_id] += 1
    return count_per_class, image_count

# === 3. Đếm ===
train_counts, train_total = count_labels(label_dir_train)
val_counts, val_total = count_labels(label_dir_val)

# === 4. Hiển thị ===
print(f"📂 Train set: {train_total} ảnh")
for cls_id in sorted(train_counts):
    print(f"  - {class_names[cls_id]}: {train_counts[cls_id]} ảnh")

print(f"\n📂 Validation set: {val_total} ảnh")
for cls_id in sorted(val_counts):
    print(f"  - {class_names[cls_id]}: {val_counts[cls_id]} ảnh")
```

```python
import matplotlib.pyplot as plt
import cv2
import glob
import os

folder_path = '/content/drive/MyDrive/yolov5/runs/val/exp'

# Lấy tất cả ảnh .png và .jpg
image_paths = sorted(glob.glob(os.path.join(folder_path, '*.png')) +
                     glob.glob(os.path.join(folder_path, '*.jpg')))

# Giới hạn số ảnh hiển thị nếu quá nhiều
max_images = 5
image_paths = image_paths[:max_images]

# Tăng kích thước figure để phóng to ảnh
fig, axes = plt.subplots(1, len(image_paths), figsize=(5 * len(image_paths), 10))

# Nếu chỉ có 1 ảnh thì axes không phải là list, cần xử lý riêng
if len(image_paths) == 1:
    axes = [axes]

for ax, img_path in zip(axes, image_paths):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(img)
    ax.set_title(os.path.basename(img_path), fontsize=12)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
import os
from collections import Counter
import matplotlib.pyplot as plt

# Đường dẫn tới thư mục chứa file label (YOLO format)
train_label_dir = '/content/drive/MyDrive/yolov5/ML/dataset_balanced/train_data/labels/train'
val_label_dir = '/content/drive/MyDrive/yolov5/ML/dataset_balanced/train_data/labels/val'

def count_classes(label_dir):
    class_counts = Counter()
    for label_file in os.listdir(label_dir):
        if label_file.endswith('.txt'):
            with open(os.path.join(label_dir, label_file), 'r') as f:
                for line in f:
                    class_id = line.strip().split()[0]
                    class_counts[class_id] += 1
    return class_counts

# Đếm class ở mỗi tập
train_counts = count_classes(train_label_dir)
val_counts = count_classes(val_label_dir)

# Tổng hợp các class xuất hiện trong cả train và val
all_classes = sorted(set(train_counts.keys()) | set(val_counts.keys()))

# Lấy số lượng theo từng class (đảm bảo có đủ)
train_values = [train_counts.get(c, 0) for c in all_classes]
val_values = [val_counts.get(c, 0) for c in all_classes]

# Vẽ biểu đồ
x = range(len(all_classes))
plt.figure(figsize=(12, 6))
plt.bar(x, train_values, width=0.4, label='Train', align='center')
plt.bar([i + 0.4 for i in x], val_values, width=0.4, label='Val', align='center')
plt.xticks([i + 0.2 for i in x], all_classes)
plt.xlabel('Class ID')
plt.ylabel('Số lượng ảnh')
plt.title('Phân bố số lượng ảnh theo class sau tăng cường (Train & Val)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
